# Proposal Idea

## 1. What is the project?

The project I have in mind is to observe active galactic nuclei (AGN) variability in dwarf galaxies (inspired from Colin Burke's talk). As discussed in his talk AGNs in dwarf galaxies have been shown to be a useful method to search for AGNs in dwarfs. 


##### 1.1. Why Photometric Data?
High-cadence photometric monitoring provides the most efficient method for detecting variability in AGN. Wide-field imaging monitors large samples with high photometric precision, allowing variability on short timescales to be measured while minimizing telescope time compared to spectroscopic observations.


There is this paper of Subaru HSC Photometry to search for AGN variablity: https://iopscience.iop.org/article/10.3847/1538-4357/ab83f3/pdf


##### 1.2. Which instrument?
This paper uses the Dark Energy Cam (DECam) at the CTIO 4 m Blanco telescope to build a catalog of high-quality optical light curves for a large sample of AGNs: https://iopscience.iop.org/article/10.3847/1538-4365/ad7d01/pdf

FOV= 2.2 degrees 

From the DECam Website (https://www.darkenergysurvey.org/the-des-project/instrument/the-camera/):

"DECam has five major sub-systems, each of which are described below:

- a 570-megapixel CCD camera,
- a low-noise electronic readout system,
- a wide-field optical corrector,
- a combination shutter-filter system, and
- a hexapod adjustor to provide stability."

##### 1.3. What Targets?
This paper provides a sample of 930 IMBH-mass AGN (ranging from $10^{4} - 10^{6.3} M_\odot$) found by searching for broad H-alpha/H-Beta in SDSS: https://arxiv.org/pdf/2503.12898 
It would be interesting to see what percent of this broadline IMBH AGN sample also show variablity.
